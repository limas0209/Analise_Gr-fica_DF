```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
food_inflation_DF = pd.read_csv("Food_Price_Inflation_DF.csv")
```

```python
food_inflation_DF
```

#GrÃ¡fico de linha da inflaÃ§Ã£o ao longo do tempo para um paÃ­s especÃ­fico

###Esse grÃ¡fico vai mostrar a evoluÃ§Ã£o da inflaÃ§Ã£o ao longo do tempo em um pÃ¡is especÃ­fico

```python
# Seleciona os dados do Indonesia
indonesia = food_inflation_DF[food_inflation_DF["country"] == "Indonesia"]


# Cria um grÃ¡fico de linha da inflaÃ§Ã£o ao longo do tempo
plt.plot(indonesia['date'], indonesia['Inflation'])
plt.xlabel('Data')
plt.ylabel('InflaÃ§Ã£o')
plt.title('InflaÃ§Ã£o de Alimentos na IndonÃ©sia')
plt.show()
```

#GrÃ¡fico em Barra: ComparaÃ§Ã£o da InflaÃ§Ã£o MÃ©dia por PaÃ­s
###Esse grÃ¡fico de barras permite comparar visualmente a inflaÃ§Ã£o mÃ©dia de alimentos entre os paÃ­ses. Podemos identificar quais paÃ­ses possuem as maiores taxas mÃ©dias de inflaÃ§Ã£o, bem como aqueles com menor inflaÃ§Ã£o, e comparar diretamente a magnitude da inflaÃ§Ã£o entre os paÃ­ses.

```python
# Agrupa os dados por paÃ­s e calcula a inflaÃ§Ã£o mÃ©dia
inflacao_media_por_pais = food_inflation_DF.groupby('country')['Inflation'].mean()

# Ordena os paÃ­ses pela inflaÃ§Ã£o mÃ©dia
inflacao_media_por_pais = inflacao_media_por_pais.sort_values(ascending=False)

# Cria um grÃ¡fico de barras
plt.bar(inflacao_media_por_pais.index, inflacao_media_por_pais.values)
plt.xlabel('PaÃ­s')
plt.ylabel('InflaÃ§Ã£o MÃ©dia')
plt.title('InflaÃ§Ã£o MÃ©dia de Alimentos por PaÃ­s')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#Histograma: DistribuiÃ§Ã£o da InflaÃ§Ã£o de Alimentos
###Esse histograma mostra a distribuiÃ§Ã£o da inflaÃ§Ã£o de alimentos, indicando quais faixas de inflaÃ§Ã£o sÃ£o mais frequentes no DataFrame. Podemos observar se a distribuiÃ§Ã£o Ã© simÃ©trica ou assimÃ©trica, se hÃ¡ outliers e qual a faixa de inflaÃ§Ã£o que concentra a maior parte dos dados.

```python
# Cria um histograma da inflaÃ§Ã£o de alimentos
plt.hist(food_inflation_DF['Inflation'].fillna(0), bins=10)
plt.xlabel('InflaÃ§Ã£o')
plt.ylabel('FrequÃªncia')
plt.title('DistribuiÃ§Ã£o da InflaÃ§Ã£o de Alimentos')
plt.show()
```

#GrÃ¡fico de Pizza: ProporÃ§Ã£o de ObservaÃ§Ãµes por Ano
###Essa anÃ¡lise usa um grÃ¡fico de pizza para visualizar a proporÃ§Ã£o de observaÃ§Ãµes de preÃ§os de alimentos no DataFrame, agrupadas por ano. Isso pode nos ajudar a entender se a frequÃªncia da coleta de dados foi consistente ao longo dos anos ou se houve variaÃ§Ãµes significativas.

```python
# Extrai o ano da coluna 'date'
food_inflation_DF['year'] = pd.to_datetime(food_inflation_DF['date']).dt.year

# Agrupa os dados por ano e conta a quantidade de observaÃ§Ãµes em cada ano
contagem_observacoes_por_ano = food_inflation_DF.groupby('year')['date'].count()

# Cria um grÃ¡fico de pizza
plt.pie(contagem_observacoes_por_ano.values, labels=contagem_observacoes_por_ano.index, autopct='%1.1f%%', startangle=90)
plt.title('ProporÃ§Ã£o de ObservaÃ§Ãµes por Ano')
plt.show()
```

# GrÃ¡fico de DispersÃ£o: RelaÃ§Ã£o entre PreÃ§o de Abertura e InflaÃ§Ã£o
###Esse grÃ¡fico de dispersÃ£o explora a relaÃ§Ã£o entre o preÃ§o de abertura dos alimentos e a inflaÃ§Ã£o. Podemos observar se hÃ¡ alguma tendÃªncia, como se preÃ§os de abertura mais altos estÃ£o associados a maior inflaÃ§Ã£o, ou se a relaÃ§Ã£o Ã© aleatÃ³ria.

```python
# Cria um grÃ¡fico de dispersÃ£o entre 'Open' e 'Inflation'
plt.scatter(food_inflation_DF['Open'], food_inflation_DF['Inflation'].fillna(0))
plt.xlabel('PreÃ§o de Abertura')
plt.ylabel('InflaÃ§Ã£o')
plt.title('RelaÃ§Ã£o entre PreÃ§o de Abertura e InflaÃ§Ã£o')
plt.show()
```
