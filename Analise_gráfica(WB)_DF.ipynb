```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
food_inflation_DF = pd.read_csv("Food_Price_Inflation_DF.csv")
```

```python
food_inflation_DF
```

#Gráfico de linha da inflação ao longo do tempo para um país específico

###Esse gráfico vai mostrar a evolução da inflação ao longo do tempo em um páis específico

```python
# Seleciona os dados do Indonesia
indonesia = food_inflation_DF[food_inflation_DF["country"] == "Indonesia"]


# Cria um gráfico de linha da inflação ao longo do tempo
plt.plot(indonesia['date'], indonesia['Inflation'])
plt.xlabel('Data')
plt.ylabel('Inflação')
plt.title('Inflação de Alimentos na Indonésia')
plt.show()
```

#Gráfico em Barra: Comparação da Inflação Média por País
###Esse gráfico de barras permite comparar visualmente a inflação média de alimentos entre os países. Podemos identificar quais países possuem as maiores taxas médias de inflação, bem como aqueles com menor inflação, e comparar diretamente a magnitude da inflação entre os países.

```python
# Agrupa os dados por país e calcula a inflação média
inflacao_media_por_pais = food_inflation_DF.groupby('country')['Inflation'].mean()

# Ordena os países pela inflação média
inflacao_media_por_pais = inflacao_media_por_pais.sort_values(ascending=False)

# Cria um gráfico de barras
plt.bar(inflacao_media_por_pais.index, inflacao_media_por_pais.values)
plt.xlabel('País')
plt.ylabel('Inflação Média')
plt.title('Inflação Média de Alimentos por País')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#Histograma: Distribuição da Inflação de Alimentos
###Esse histograma mostra a distribuição da inflação de alimentos, indicando quais faixas de inflação são mais frequentes no DataFrame. Podemos observar se a distribuição é simétrica ou assimétrica, se há outliers e qual a faixa de inflação que concentra a maior parte dos dados.

```python
# Cria um histograma da inflação de alimentos
plt.hist(food_inflation_DF['Inflation'].fillna(0), bins=10)
plt.xlabel('Inflação')
plt.ylabel('Frequência')
plt.title('Distribuição da Inflação de Alimentos')
plt.show()
```

#Gráfico de Pizza: Proporção de Observações por Ano
###Essa análise usa um gráfico de pizza para visualizar a proporção de observações de preços de alimentos no DataFrame, agrupadas por ano. Isso pode nos ajudar a entender se a frequência da coleta de dados foi consistente ao longo dos anos ou se houve variações significativas.

```python
# Extrai o ano da coluna 'date'
food_inflation_DF['year'] = pd.to_datetime(food_inflation_DF['date']).dt.year

# Agrupa os dados por ano e conta a quantidade de observações em cada ano
contagem_observacoes_por_ano = food_inflation_DF.groupby('year')['date'].count()

# Cria um gráfico de pizza
plt.pie(contagem_observacoes_por_ano.values, labels=contagem_observacoes_por_ano.index, autopct='%1.1f%%', startangle=90)
plt.title('Proporção de Observações por Ano')
plt.show()
```

# Gráfico de Dispersão: Relação entre Preço de Abertura e Inflação
###Esse gráfico de dispersão explora a relação entre o preço de abertura dos alimentos e a inflação. Podemos observar se há alguma tendência, como se preços de abertura mais altos estão associados a maior inflação, ou se a relação é aleatória.

```python
# Cria um gráfico de dispersão entre 'Open' e 'Inflation'
plt.scatter(food_inflation_DF['Open'], food_inflation_DF['Inflation'].fillna(0))
plt.xlabel('Preço de Abertura')
plt.ylabel('Inflação')
plt.title('Relação entre Preço de Abertura e Inflação')
plt.show()
```
